Primero limpio el texto de caracteres que no sean letras, de esta forma queda mas legible y se pueden seleccionar mejor las palabras.

```python
def clear_text (text):
    str_chain = text.replace("<","").replace("!","").replace(">","").replace("("," ").replace(")"," ").replace("["," ").replace("]"," ").replace("-"," ").replace("*"," ").replace(":"," ")
    str_chain = str_chain.replace("0"," ").replace("1"," ").replace("2"," ").replace("3"," ").replace("4"," ").replace("5"," ").replace("6"," ").replace("7"," ").replace("8"," ").replace("9"," ")
    str_chain = str_chain.replace("+"," ").replace("/"," ").replace("."," ").replace(","," ").replace("@"," ").replace(" \ "," ").replace("`"," ").replace("´"," ")
    str_chain = str_chain.replace("?"," ").replace("$"," ").replace("="," ").replace("%"," ").replace("&"," ").replace (' " ' , "")
    return str_chain
```

Despues de limpiarlo, separo en una lista cada uno de los saltos de linea presentes en el texto

```python
def line_split (text):
    list = text.split('\n')
    list = [elem for elem in list if elem.strip()]
    return list
```

Ahora, usando el metodo split, separo en una lista las palabras de cada salto de linea, esto me permite hacer un acceso al segundo elemento y verificar si comienza con una vocal o no, en caso de que la lista tenga un solo elemento, me posiciono en la posicion de dicho elemento, de esta forma no queda nada en el texto sin verificar.

```python
def vowels_check (list):
    vowels = ['a','e','i','o','u','A','E','I','O','U']
    for text in list:
        words = text.split()
        if (len (words) > 1):
            check = words [1]
        else:
            check = words[0]
        for vowel in vowels:
            if (vowel == check[0]):
                print(text)
```
