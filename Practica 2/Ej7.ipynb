7. Dada una frase contar mayúsculas, minúsculas, caracteres no letras y la cantidad de
palabras sin distinguir entre mayúsculas y minúsculas.
text = """ La brecha salarial alcanzó el 27,7%: las mujeres ocupadas
debieron trabajar 8 días y 10 horas más que los varones ocupados para
ganar lo mismo que ellos en un mes. """

Importando string, la funcion lower_letters recibe el texto (text) como parametro, la variable lowers recibe de stirng todos los caracteres en minuscula, sumado a los que son minuscula y poseen tilde.
La cariable low cuenta la cantidad de caracteres que son minusculas.
El for itera las letras en el texto, si la letra se encuentra en lowers, se le suma uno a low

```python
import string
def lower_letters (text):
    lowers = string.ascii_lowercase + 'áéíóú'
    low = 0
    for letter in text:
        if letter in lowers:
            low += 1
    print(f'La cantidad de minusculas es: {low}')
```

La funcion upper_letters tiene la misma funcion que la vista previamente, cambia el hecho de que esta cuenta cuantas mayusculas hay en el texto

```python
def upper_letters (text):
    uppers = string.ascii_uppercase + 'ÁÉÍÓÚ'
    up = 0
    for letter in text:
        if letter in uppers:
            up += 1
    print(f'La cantidad de mayusculas es: {up}')
```

En la funcion non_letters, recibe el texto como parametro y analiza que caracteres no son alfabeticos.
nolet recibe de string los caracteres que son digitos y los que son de puntuacion
El for itera las letras, si la letra esta dentro de nolet, se le suma uno a no.

```python
def non_letters (text):
    nolet = string.digits + string.punctuation
    no = 0
    for letter in text:
        if letter in nolet:
            no +=1
    print(f'La cantidad de caracteres no alfabeticos es {no}')
```

Programa principal con el texto y las funciones llamadas

```python
text = """ La brecha salarial alcanzó el 27,7%: las mujeres ocupadas
debieron trabajar 8 días y 10 horas más que los varones ocupados para
ganar lo mismo que ellos en un mes. """

lower_letters(text)
upper_letters(text)
non_letters(text)
```
