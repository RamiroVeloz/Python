Si ahora queremos saber si contiene la letra a y también la letra n, ¿cómo lo
modificamos?


Como modificacion, solo agrego una condicion en el if, dado que el ejercicio indica que se debe analizar si en una palabra existe la letra a Y la letra n

```python
word = input("Ingresá una palabra: ")
if "a" and "n" in word:
    print("Hay al menos una letra a y una letra n.")
else:
    print("No hay letras a y n en la misma palabra. ")
```
