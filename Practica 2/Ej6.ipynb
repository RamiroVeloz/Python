6. Dada una frase y un string ingresados por teclado (en ese orden), genere una lista de
palabras, y sobre ella, informe la cantidad de palabras en las que se encuentra el
string. No distingir entre mayúsculas y minúsculas

Para este ejercicio, implemento un metodo check_string, el cual recibe el texto (text) y una opcion (op).
Utilizando la funcion Counter del paquete collections, cuento cada una de las palabras que estan en el texto.
Previo a contarlas en el for, hago una lista donde separo las palabras y las convierto en minusculas con lower. Separarlas dado que no puedo iterar las palabras una a una dentro del for, si hiciera for word in text, estaria iterando las letras. De esta forma separo las palabras y puedo trabajarlas como elementos individuales.
Si la palabra esta dentro del diccionario que me devuelve el metodo counter (diccionario que se encuentra en ctr), devuelvo la palabra y su valor, si no, devuelvo 0

```python
from collections import Counter

def check_string (text,op):
    ctr = Counter()
    list = text.lower().split()
    for word in list:
        ctr [word] +=1
    if (op in ctr):
        return ctr[op]
    else:
        return 0
```

```python
text = """Tres tristes tigres, tragaban trigo en un trigal, en tres tristes
trastos, tragaban trigo tres tristes tigres."""

op = input ('Ingrese una palabra para buscar en la frase: ')
value = check_string(text,op)
print (f'La palabra op tiene un total de {value} apariciones')
```
