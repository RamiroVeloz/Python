3. Dado el siguiente texto guardado en la varible jupyter_info, solicite por teclado una
letra e imprima las palabras que contengan dicha letra. En caso que no se haya
ingresado un letra, indique el error.

Para este ejercicio, creo una funcion llamada letter_count que recibe como parametros el texto (text) y una opcion (choice), lo que hace este metodo es devolver la variable interna cant, que con el metodo count, cuenta las ocurrencias de choice dentro del texto.

```python
def letter_count (text,choice):
    cant = text.count(choice)
    return cant
```

Programa principal con el texto a analizar, choice recibiendo una entrada por teclado y cant recibiendo la cantidad de ocurrencias, que luego se imprimen en el print final

```python
jupyter_info = """ JupyterLab is a web-based interactive development
environment for Jupyter notebooks, code, and data. JupyterLab is
flexible: configure and arrange the user interface to support a wide
range of workflows in data science, scientific computing, and machine
learning. JupyterLab is extensible and modular: write plugins that add
new components and integrate with existing ones. """

choice = input('Ingrese una letra:')

cant = letter_count(jupyter_info,choice)

print(f'La cantidad de ocurrencias que tiene la letra {choice} en el texto es: {cant}') 
```
