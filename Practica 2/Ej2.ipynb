Igual que en el ejercicio anterior, limpio el texto de caracteres innecesarios, e importo el metodo counter para mas adelante

```python
from collections import Counter

def clear_text (text):
    str_chain = text.replace("<","").replace("!","").replace(">","").replace("("," ").replace(")"," ").replace("["," ").replace("]"," ").replace("-"," ").replace("*"," ").replace(":"," ")
    str_chain = str_chain.replace("0"," ").replace("1"," ").replace("2"," ").replace("3"," ").replace("4"," ").replace("5"," ").replace("6"," ").replace("7"," ").replace("8"," ").replace("9"," ")
    str_chain = str_chain.replace("+"," ").replace("/"," ").replace("."," ").replace(","," ").replace("@"," ").replace(" \ "," ").replace("`"," ").replace("´"," ")
    str_chain = str_chain.replace("?"," ").replace("$"," ").replace("="," ").replace("%"," ").replace("&"," ").replace (' " ' , "")
    return str_chain
```

Como en el caso anterior, separo los elementos del texto en una lista, esta vez por palabras, luego los vuelvo a separar con la condicion de que a los elementos se les aplique el metodo lower que transforma todas las palabras del string en minuscula.

```python
def line_split (text):
    list = text.split()
    list = [elem for elem in list if elem.strip()]
    list = [elem.lower() for elem in list]
    return list
```

En esta funcion, uso el metodo counter para contar cada una de las palabras dentro del texto

```python
def max_word (list):
    count = Counter()
    for elem in list:
        count[elem]  += 1
    #print(count)
    print (f'La palabra mas comun fue {count.most_common(1)[0][0]} con un total de {count.most_common(1)[0][1]} ocurrencias')
```
