9. Escriba un programa que solicite por teclado una palabra y calcule el valor de la misma
dada la siguiente tabla de valores del juego Scrabble:
Letra valor
A, E, I, O, U, L, N, R, S, T = 1
D, G  = 2
B, C, M, P = 3
F, H, V, W, Y = 4
K = 5
J, X = 8
Q, Z = 10


Imporot string para usarlo mas adelante.
La funcion calculate_score genera un diccionario con los valores que tiene cada letra dentro del juego.
Si la letra que ingresa como parametro dentro de la funcion, se encuentra dentro del diccionario, se devolvera el punto correspondiente.

```python
import string

def calculate_score (letter):
    scores = {'A' : 1,'E' : 1,'I': 1,'O' : 1,'U' : 1,'N' : 1,'R' : 1,'S' : 1,'T' : 1,
    'D' : 2,'G' : 2,
    'B' : 3,'C' : 3,'M' : 3,'P' : 3,
    'F' : 4,'H' : 4,'V' : 4,'W' : 4,'Y' : 4,
    'K' : 5,
    'J' : 8,'X' : 8,
    'Q' : 10,'Z' : 10}

    point = 0

    if (letter in scores):
        point = scores[letter]
    
    return point
```

Esta funcion analiza la palabra (text) para calcular los puntos.
Se genera una variable que tiene todos los caracteres que no son alfabeticos. 
El for itera las letras del texto y, si no sucede que la letra esta dentro de los caracteres no alfabeticos, se llama a una variable auxiliar que almacena el valor que devuelve la funcion calculate_score, la cual recibe la letra analizada. luego se devuelven los puntos de la letra.

```python
def analize_text (text):
    points = 0
    not_char = string.digits + string.punctuation + ' '
    for letter in text:
        letter = letter.upper()
        if not (letter in not_char):
            aux = calculate_score(letter)
            points += aux
    return points
```

Programa principal, text recibe la palabra por teclado, la variable point llama a la funcion analize_text que recibe la palabra y devuelve los puntos que vale cada letra, luego se imprimen en pantalla.

```python
text = input ('Ingrese una palabra: ')
points = analize_text(text)
print(f'El puntaje de la palabra ingresada es: {points}')
```
